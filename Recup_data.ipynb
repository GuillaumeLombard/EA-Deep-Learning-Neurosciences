# üß† R√©sum√© : Matrices SC et FC

## 1Ô∏è‚É£ Matrice SC : **Structural Connectivity (Connectivit√© Structurelle)**

La matrice SC repr√©sente la connectivit√© anatomique entre les r√©gions c√©r√©brales.  
Elle est construite √† partir des donn√©es de diffusion (DSI) via la tractographie, en suivant les fibres de substance blanche.

- **SC(i, j)** : Nombre de fibres reliant les r√©gions i et j, normalis√© par la longueur et la surface des r√©gions.  
- **Interpr√©tation** : Plus SC(i, j) est √©lev√©e, plus les r√©gions sont structurellement connect√©es.  
- **Source** : Imagerie de diffusion (DSI).  
- **Nature** : Fixe dans le temps (anatomie).

---

## 2Ô∏è‚É£ Matrice FC : **Functional Connectivity (Connectivit√© Fonctionnelle)**

La matrice FC repr√©sente la synchronisation fonctionnelle entre les r√©gions c√©r√©brales.  
Elle est obtenue √† partir de l'IRM fonctionnelle de repos (rs-fMRI) en calculant la corr√©lation temporelle des signaux BOLD.

- **FC(i, j)** : Coefficient de corr√©lation de Pearson entre les s√©ries temporelles des r√©gions i et j.  
- **Interpr√©tation** : Plus FC(i, j) est proche de 1, plus les r√©gions sont fonctionnellement synchronis√©es.  
- **Source** : IRM fonctionnelle (BOLD fMRI).  
- **Nature** : Dynamique et variable dans le temps.

## üéØ Application dans l'√©tude SCHZ vs CTRL

La schizophr√©nie est associ√©e √† des alt√©rations de la connectivit√© fonctionnelle (FC), m√™me lorsque la connectivit√© structurelle (SC) reste relativement stable.  
Comparer ces matrices entre les groupes permet de mettre en √©vidence des **biomarqueurs potentiels** li√©s √† la maladie.

**SC et FC ensemble** offrent une vision compl√®te de la connectivit√© c√©r√©brale, en int√©grant **anatomie** et **dynamique fonctionnelle**. üß†‚ú®


## **Pistes d'analyse en Deep Learning pour le projet SCHZ/CTRL**

### 1Ô∏è‚É£ Classification SCHZ vs CTRL
- **Objectif** : Classer les sujets en SCHZ ou CTRL √† partir des matrices SC et FC.
- **M√©thode** : CNN 2D pour traiter les matrices comme des images, extraire des caract√©ristiques spatiales et r√©aliser une classification binaire.
- **Id√©e bonus** : Comparer les performances selon les r√©solutions et fusionner SC et FC pour am√©liorer la pr√©cision.

---

### 2Ô∏è‚É£ Apprentissage contrastif (Siamese Network)
- **Objectif** : Apprendre √† diff√©rencier les connectomes SCHZ et CTRL par paires.
- **M√©thode** : R√©seau Siamois (deux CNN partageant les poids) pour mesurer la similarit√© entre deux matrices.
- **Int√©r√™t** : Identifier les r√©gions les plus discriminantes en observant les paires les plus dissemblables.

---

### 3Ô∏è‚É£ R√©duction de dimension et Clustering
- **Objectif** : R√©duire la dimension des matrices et identifier des groupes distincts (SCHZ vs CTRL).
- **M√©thode** : Autoencodeur pour compresser les matrices, suivi d'un clustering (K-Means) dans l'espace latent.
- **Bonus** : Visualiser les groupes selon les r√©solutions et types de matrices (ADC, gFA, etc.).

---

### 4Ô∏è‚É£ Analyse par Graph Neural Networks (GNN)
- **Objectif** : Traiter les connectomes comme des graphes et capturer les relations entre r√©gions.
- **M√©thode** : GCN (Graph Convolutional Network) pour extraire des motifs structurels et classifier SCHZ vs CTRL.
- **Atout** : Exploiter la nature "r√©seau" des connectomes c√©r√©braux.

---

### 5Ô∏è‚É£ Analyse d‚Äôinterpr√©tabilit√©
- **Objectif** : Identifier les connexions ou r√©gions c√©r√©brales les plus importantes pour la classification.
- **M√©thode** : Grad-CAM pour CNN ou Attention Maps pour GNN.
- **R√©sultat** : Visualiser les r√©gions c√©r√©brales cl√©s pour la distinction SCHZ/CTRL.

---

### 4.1 / IXC

```python
file_path = 'Data/labels_index_CORTICAL_Laus2008_all_scales.mat'

# Charger le fichier
labels_data = sio.loadmat(file_path)

# Extraire les indices corticaux
ixc = labels_data['ixc']  # Contient les indices pour toutes les √©chelles

# Afficher les indices pour chaque √©chelle
for i, indices in enumerate(ixc[0]):  # ixc est un tableau de tableaux
    print(f"Indices corticaux pour l'√©chelle {i+1} :")
    print(indices[0])  # Affiche les indices d'une √©chelle
    print("-" * 50)
```

Le tableau `ixc` extrait de `labels_index_CORTICAL_Laus2008_all_scales.mat` contient les **indices des r√©gions corticales** pour diff√©rentes **r√©solutions** de l'atlas Lausanne2008 (83, 129, 234, 463, 1015 r√©gions).

- Chaque tableau dans `ixc` correspond √† une **√©chelle sp√©cifique**.
- Les indices list√©s repr√©sentent **uniquement les r√©gions corticales**, en excluant les r√©gions sous-corticales, le tronc c√©r√©bral et certaine r√©gions corticales frontali√®res.
- Par exemple, pour l'√©chelle 83 r√©gions, `ixc` contient 68 indices (les 68 r√©gions corticales).

Il :
- Permet de **filtrer les matrices SC et FC** pour ne garder que les connexions corticales.
- Est utile pour analyser sp√©cifiquement le **cortex c√©r√©bral** sans inclure les structures sous-corticales.


```python
# Charger les indices corticaux
labels_file = 'Data/labels_index_CORTICAL_Laus2008_all_scales.mat'
labels_data = sio.loadmat(labels_file)
cortical_indices_83 = labels_data['ixc'][0, 0].flatten() - 1  # Indices pour l'√©chelle 83

# Charger les matrices SC et FC depuis SC_FC_Connectomes
file_path = 'Data/27_SCHZ_CTRL_dataset.mat'
with h5py.File(file_path, 'r') as f:
    # Acc√©der aux matrices SC et FC (exemple avec SC_density et FC_correlation)
    SC_ctrl = f['SC_FC_Connectomes/SC_density']['ctrl'][0, 0]
    SC_schz = f['SC_FC_Connectomes/SC_density']['schz'][0, 0]
    FC_ctrl = f['SC_FC_Connectomes/FC_correlation']['ctrl'][0, 0]
    FC_schz = f['SC_FC_Connectomes/FC_correlation']['schz'][0, 0]

    # Charger les matrices
    SC_ctrl = f[SC_ctrl][:]
    SC_schz = f[SC_schz][:]
    FC_ctrl = f[FC_ctrl][:]
    FC_schz = f[FC_schz][:]
    
    # Appliquer le masque cortical aux matrices
    SC_ctrl_cortical = SC_ctrl[:, cortical_indices_83, :][:, :, cortical_indices_83]
    SC_schz_cortical = SC_schz[:, cortical_indices_83, :][:, :, cortical_indices_83]
    FC_ctrl_cortical = FC_ctrl[:, cortical_indices_83, :][:, :, cortical_indices_83]
    FC_schz_cortical = FC_schz[:, cortical_indices_83, :][:, :, cortical_indices_83]

# Afficher les nouvelles formes
print(f"SC CTRL cortical shape: {SC_ctrl_cortical.shape}")
print(f"SC SCHZ cortical shape: {SC_schz_cortical.shape}")
print(f"FC CTRL cortical shape: {FC_ctrl_cortical.shape}")
print(f"FC SCHZ cortical shape: {FC_schz_cortical.shape}")
```

```python
# Chemins des fichiers
file_path = 'Data/27_SCHZ_CTRL_dataset.mat'
labels_path = 'Data/labels_index_CORTICAL_Laus2008_all_scales.mat'

# Charger les indices corticaux
labels_data = sio.loadmat(labels_path)
cortical_indices = labels_data['ixc'][0, 0].flatten() - 1  # MATLAB indexation

# Charger les matrices FC
with h5py.File(file_path, 'r') as f:
    FC_ctrl = f['SC_FC_Connectomes/FC_correlation/ctrl'][0, 0]  # R√©f√©rence
    FC_schz = f['SC_FC_Connectomes/FC_correlation/schz'][0, 0]  # R√©f√©rence

    FC_ctrl_data = f[FC_ctrl][:]
    FC_schz_data = f[FC_schz][:]

# Appliquer le masque cortical
FC_ctrl_cortical = FC_ctrl_data[:, cortical_indices, :][:, :, cortical_indices]
FC_schz_cortical = FC_schz_data[:, cortical_indices, :][:, :, cortical_indices]

# Moyennes pour visualisation
mean_FC_ctrl = np.mean(FC_ctrl_data, axis=0)
mean_FC_schz = np.mean(FC_schz_data, axis=0)
mean_FC_ctrl_cortical = np.mean(FC_ctrl_cortical, axis=0)
mean_FC_schz_cortical = np.mean(FC_schz_cortical, axis=0)

# Affichage
fig, ax = plt.subplots(2, 2, figsize=(12, 10))

ax[0, 0].imshow(mean_FC_ctrl, cmap='plasma')
ax[0, 0].set_title('FC CTRL (83x83) - Avant filtre')

ax[0, 1].imshow(mean_FC_schz, cmap='plasma')
ax[0, 1].set_title('FC SCHZ (83x83) - Avant filtre')

ax[1, 0].imshow(mean_FC_ctrl_cortical, cmap='plasma')
ax[1, 0].set_title('FC CTRL (68x68) - Apr√®s filtre')

ax[1, 1].imshow(mean_FC_schz_cortical, cmap='plasma')
ax[1, 1].set_title('FC SCHZ (68x68) - Apr√®s filtre')

plt.tight_layout()
plt.show()
```

Les matrices filtr√©es conservent les principales connexions corticales et offrent une meilleure base pour appliquer des algorithmes de deep learning. Voici les principales raisons : 

- **Focus sur le cortex c√©r√©bral** : Exclusion des r√©gions sous-corticales et du tronc c√©r√©bral.
- **R√©duction de la dimensionnalit√©** : Moins de complexit√©, meilleure g√©n√©ralisation.
- **Moins de bruit** : √âlimine les r√©gions moins pertinentes pour l'√©tude de la schizophr√©nie.
- **Alignement avec la litt√©rature** : Comparaison plus facile avec les √©tudes existantes.
